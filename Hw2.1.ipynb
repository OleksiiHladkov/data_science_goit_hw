```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"**

```python
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
table_name = 'Коефіцієнт народжуваності в регіонах України'
# data = pd.read_html(url, match=table_name, thousands="", decimal=".")
data = pd.read_html(url, match=table_name, thousands=' ', decimal=',')
br_data_1 = data[0]
br_data_1
```

- Вивести перші рядки таблиці за допомогою методу head

```python
br_data_1.head()
```

- Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
print('rows:', br_data_1.shape[0], 'columns:', br_data_1.shape[1])
```

- Замініть у таблиці значення "—" на значення NaN

```python
br_data_2 = br_data_1.replace('—', np.nan)
br_data_2
```

- Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
br_data_2.dtypes
```

- Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
# before
print('type of column "2014" before:', br_data_2['2014'].dtype)
print('type of column "2019" before:', br_data_2['2019'].dtype)

br_data_3 = br_data_2.copy()

# after
br_data_3['2014'] = pd.to_numeric(br_data_3['2014'])
br_data_3['2019'] = pd.to_numeric(br_data_3['2019'])
print('type of column "2014" after:', br_data_3['2014'].dtype)
print('type of column "2019" after:', br_data_3['2019'].dtype)
```

- Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
print("Number of NaN:")
for i in br_data_3.items():
  print(f'column "{i[0]}": ', br_data_3[i[0]].isnull().sum())
```

- Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
print('rows before: ', br_data_3.shape[0])

br_data_4 = br_data_3.copy()

index = br_data_4.shape[0]-1
br_data_4 = br_data_4.drop([index])
print('rows after: ', br_data_4.shape[0])
br_data_4
```

- Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
br_data_5 = br_data_4.copy()

new_dict = {}

for i in br_data_4.items():
  col_name = i[0]
  if br_data_4[col_name].isnull().sum() == 0:
    continue
  new_dict[col_name] = round(br_data_5[col_name].mean(), 1)

br_data_5 = br_data_5.fillna(new_dict)
br_data_5
```

- Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
mean_2019 = br_data_5['2019'].mean()
br_data_5.query(f'`2019` > {mean_2019}')['Регіон'].tolist()
```

- У якому регіоні була найвища народжуваність у 2014 році?

```python
max_2014 = br_data_5['2014'].max()
br_data_5.query(f'`2014` == {max_2014}')['Регіон'].tolist()
```

- Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
br_2019 = pd.DataFrame()
br_2019['Регіон'] = br_data_5['Регіон']
br_2019['2019'] = br_data_5['2019']

br_2019.plot(x='Регіон', kind='bar', title='Birth rate in 2019')
plt.show()
```

# **Візуалізація**

```python
import matplotlib.pyplot as plt

df = br_data_5.query('Регіон in ["Луганська", "Закарпатська", "Крим", "Чернігівська", "Київська"]')
df
```

```python
plt.figure(figsize=(14,8))

colums_count = df.shape[1]
columns = df.columns.to_list()[1:colums_count]

for index, row in df.iterrows():
  plt.plot(columns, row.values[1:colums_count], label=row.values[0], marker="D")

plt.title('Динаміка коефіцієнту народжуваності по областях', fontsize=15)
plt.xlabel('Рік', fontsize=10, color='midnightblue')
plt.ylabel('Коефіцієнт', fontsize=10, color='midnightblue')
plt.legend()
plt.show()
```

```python
df['2014'].to_list()

plt.figure(figsize=(10,5))

plt.bar(df['Регіон'].to_list(), df['2014'].to_list())

plt.xlabel('Область', fontsize=10, color="midnightblue")
plt.ylabel('Коефіцієнт', fontsize=10, color="midnightblue")
plt.title('Коефіцієнт народжуваності по областях у 2014 році', fontsize=15)
plt.show()
```
