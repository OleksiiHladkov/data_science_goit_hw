- Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

```python
labels_list = []
for i in data.items():
  column_name = i[0]
  if data[column_name].isnull().sum() != 0 and column_name != "Мова програмування":
    labels_list.append(column_name)

data_1 = data.drop(labels_list, axis=1)
data_1.head()
```

- Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
python_data = data_2.query('`Мова програмування`=="Python"')
python_data.head()
```

- Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
python_data_gb = python_data.groupby('Посада')
python_data_gb.head()
```

- Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
python_salary_data = python_data[['Посада', 'Зарплата на місяць']]
python_salary_result = python_salary_data.groupby('Посада').agg(['min', 'max'])
python_salary_result
```

- Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
  # знаходимо всі записи по поточній посаді з фрейму до группування
  new_df = python_salary_data.query(f'Посада == "{row.name}"')
  # знаходимо середню по цим записам
  avg = new_df['Зарплата на місяць'].mean()

  return [row[0], row[1], avg]

python_salary_result['Зарплата на місяць','avg'] = 0
python_salary_result = python_salary_result.apply(fill_avg_salary,  axis=1, result_type='broadcast')
python_salary_result
```

- Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
python_salary_result['Зарплата на місяць','avg'].describe()
```

# **Візуалізація**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.lineplot(data=python_salary_result)
# sns.set_theme(rc={'figure.figsize':(10.0,5.0)})

plt.xlabel('Посада', size=16, family='monospace')
plt.ylabel('Заробітна плата', size=16, family='monospace')
plt.title('Заробітна плата Python, USD', size=20,)

plt.show()
```

```python
python_data_exp = python_data[['Посада', 'Загальний досвід роботи', 'Зарплата на місяць']]
python_data_exp = python_data_exp.replace(to_replace='10 і більше років', value='10')
python_data_exp = python_data_exp.replace(to_replace='менше 3 місяців', value='0.25')
python_data_exp['Загальний досвід роботи'] = pd.to_numeric(python_data_exp['Загальний досвід роботи'])

sns.scatterplot(data=python_data_exp, x='Загальний досвід роботи', y='Зарплата на місяць')

plt.xlabel('Загальний досвід роботи, років', size=16, family='monospace')
plt.ylabel('Заробітна плата, USD', size=16, family='monospace')
plt.title('Залежність заробітної плати від досвіду роботи в Python', size=20)
plt.xticks(np.arange(11))

plt.show()
```

```python
sns.stripplot(x='Посада', y='Зарплата на місяць', hue='Загальний досвід роботи', data=python_data_exp)

plt.xlabel('Посада', size=16, family='monospace')
plt.ylabel('Заробітна плата, USD', size=16, family='monospace')
plt.title('Залежність заробітної плати від досвіду роботи по посадах в Python', size=20)

plt.show()
```
